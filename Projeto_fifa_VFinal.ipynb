#### Como podemos verificar perdemos muito da nossa base de dados se utilizarmos esse metédo, por conta disso iremos verificar onde estão os dados faltantes e vamos preenche-los da melhor forma possível. 

```python
faltantes = base.isnull().sum().sort_values(ascending=False)[:45]

faltantes
```

##### Podemos ver que os dados faltantes são referentes a seleção, posição e clube.
#### Sabendo que nem todos os jogadores vestiram a camisa de suas seleções, nem todos tem uma estimativa para todas as posições
#### e que alguns jogadores ficam sem clube, podemos assumir que esses valores serão zero. 

```python
base['national_team'].fillna('Nenhum', inplace = True)
base['national_jersey_number'].fillna('0', inplace = True)
base['national_team_position'].fillna('Nenhuma', inplace = True)
base['national_rating'].fillna('0', inplace = True)
base['tags'].fillna('Nenhuma', inplace = True)
base['traits'].fillna('Nenhuma', inplace = True)

base['RAM'].fillna('0', inplace = True)
base['LM'].fillna('0', inplace = True)
base['CAM'].fillna('0', inplace = True)
base['RW'].fillna('0', inplace = True)
base['RF'].fillna('0', inplace = True)
base['CF'].fillna('0', inplace = True)
base['LF'].fillna('0', inplace = True)
base['LW'].fillna('0', inplace = True)
base['RS'].fillna('0', inplace = True)
base['ST'].fillna('0', inplace = True)
base['LCM'].fillna('0', inplace = True)
base['CM'].fillna('0', inplace = True)
base['RCM'].fillna('0', inplace = True)
base['RM'].fillna('0', inplace = True)
base['LWB'].fillna('0', inplace = True)
base['LDM'].fillna('0', inplace = True)
base['LS'].fillna('0', inplace = True)
base['CDM'].fillna('0', inplace = True)
base['RDM'].fillna('0', inplace = True)
base['RWB'].fillna('0', inplace = True)
base['LB'].fillna('0', inplace = True)
base['LCB'].fillna('0', inplace = True)
base['CB'].fillna('0', inplace = True)
base['RCB'].fillna('0', inplace = True)
base['RB'].fillna('0', inplace = True)
base['LAM'].fillna('0', inplace = True)

base['club_jersey_number'].fillna('0', inplace = True)
base['club_position'].fillna('Nenhuma', inplace = True)
base['club_rating'].fillna('0', inplace = True)
base['club_team'].fillna('Sem clube', inplace = True)


faltantes2 = base.isnull().sum().sort_values(ascending=False)[:10]

faltantes2
```

#### Agora para os dados faltantes das colunas 'contract_end_year', 'value_euro', 'release_clause_euro' e 'wage_euro' vamos tratar da seguinte forma:
#### Preenchendos com os valores de suas medianas.

```python
release_clause_euro_mediana = base.release_clause_euro.median()
base['release_clause_euro'].fillna(release_clause_euro_mediana, inplace = True)

value_euro_mediana = base.value_euro.median()
base['value_euro'].fillna(value_euro_mediana, inplace = True)

wage_euro_mediana = base.wage_euro.median()
base['wage_euro'].fillna(wage_euro_mediana, inplace = True)

base['contract_end_year'].fillna(2023, inplace = True)

base['club_join_date'].fillna(2018-1-1, inplace = True)

```
